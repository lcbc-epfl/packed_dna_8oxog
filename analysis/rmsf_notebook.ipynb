```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

This notebook describes the rmsf values per base for three different classical MD simulations: wt NCP, 8oxoguanine defect in region 1 (facing histone core), 8oxoguanine defect region 2 (facing solvent).

RMSF values were calculated in the following way:
1. RMSD values were calculated for each system across the trajectory (compared to an average structure)
2. The frame with the lowest RMSD value was selected as reference for each system
3. The trajectory for each system was aligned with its selected reference frame
4. The gromacs rmsf function was used with the selected reference frame to calculate the rmsf value per base ID
5. RMSF values per base ID for the three systems were plotted here by section of 21 base pairs (42 bases total)

```python
df_wt_res_sec1=pd.read_csv('./wt_res_sec1_rmsf.csv')
df_wt_res_sec2=pd.read_csv('./wt_res_sec2_rmsf.csv')
df_wt_res_sec3=pd.read_csv('./wt_res_sec3_rmsf.csv')
df_wt_res_sec4=pd.read_csv('./wt_res_sec4_rmsf.csv')
df_r1_res_sec1=pd.read_csv('./r1_res_sec1_rmsf.csv')
df_r1_res_sec2=pd.read_csv('./r1_res_sec2_rmsf.csv')
df_r1_res_sec3=pd.read_csv('./r1_res_sec3_rmsf.csv')
df_r1_res_sec4=pd.read_csv('./r1_res_sec4_rmsf.csv')
df_r2_res_sec1=pd.read_csv('./r2_res_sec1_rmsf.csv')
df_r2_res_sec2=pd.read_csv('./r2_res_sec2_rmsf.csv')
df_r2_res_sec3=pd.read_csv('./r2_res_sec3_rmsf.csv')
df_r2_res_sec4=pd.read_csv('./r2_res_sec4_rmsf.csv')
```

```python
x_wt_sec1 = list(df_wt_res_sec1.iloc[:,0].astype(str))
y_wt_sec1 = list(df_wt_res_sec1.iloc[:,1])
x_r1_sec1 = list(df_r1_res_sec1.iloc[:,0].astype(str))
y_r1_sec1 = list(df_r1_res_sec1.iloc[:,1])
x_r2_sec1 = list(df_r2_res_sec1.iloc[:,0].astype(str))
y_r2_sec1 = list(df_r2_res_sec1.iloc[:,1])
bases = {'x_wt_sec1', 'x_r1_sec1','x_r2_sec1'}
rmsf_values = {'y_wt_sec1', 'y_r1_sec1', 'y_r2_sec1'}
plt.rcParams["figure.figsize"] = [15.50, 3.50]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
x_axis_sec1 = np.arange(len(x_wt_sec1))
plt.bar(x_axis_sec1 - 0.2, y_wt_sec1, 0.2, label = 'Wild Type', color = 'blue')
plt.bar(x_axis_sec1 + 0.2, y_r1_sec1, 0.2, label = '8oxoG Region 1', color = 'green')
plt.bar(x_axis_sec1 + 0, y_r2_sec1, 0.2, label = '8oxoG Region 2', color = 'orange')
plt.xlabel('Base ID')
plt.ylabel('rmsf (A)')
plt.title('RMSF Section 1 Values')
plt.xticks(x_axis_sec1, x_wt_sec1)
plt.ylim([0,0.3])
plt.legend()
plt.show()
```
